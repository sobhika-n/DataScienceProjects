### Create name and race columns

```python
fake = Faker()
# Generate fake names and races
data['Name'] = [fake.name() for _ in range(len(data))]
data['Race'] = [fake.random_element(elements=('White', 'Black', 'Asian', 'Hispanic')) for _ in range(len(data))]
```
